```python
def fun(a):
    if a < 21:
        return "Kid"
    elif (a >= 21) and (a < 60):
        return "Adult"
    else:
        return "Old"
    
titanic_data['Age_Status'] = titanic_data.Age.map(fun)
```

```python
titanic_data
```

```python
def incr_fare(r):
    if (r['Sex'] == 'male') and (r['Age_Status'] == 'Adult'):
        return 2*r['Fare']
    else:
        return r['Fare']
    
titanic_data['Fare'] = titanic_data.apply(incr_fare, axis=1)
```
